## 2.对标签One-Hot编码

```python
from torch.nn.functional import one_hot
import torch

x=torch.tensor(x_train,dtype=torch.float)
y=one_hot(torch.tensor(y_train_label,dtype=torch.long).long(), num_classes=10)
print(y[0])
"""
对于使用交叉熵损失函数，不需要对标签进行one-hot编码，直接使用原始标签即可
下面步骤中，直接使用原始标签即可，不需要对标签进行one-hot编码
"""
```

## 3.模型定义

```python
import torch.nn as nn
class NeuralNetwork(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = torch.nn.Flatten()
        self.linear_relu_stack = torch.nn.Sequential(
            nn.Linear(28*28, 312),
            nn.ReLU(),
            nn.Linear(312, 256),
            nn.ReLU(),
            nn.Linear(256, 10),
        )
    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits

#model=NeuralNetwork()
```

## 6.训练循环

```python
from tqdm import tqdm

batch_size=320
epochs=1024

device='cpu'

model=NeuralNetwork().to(device)
loss_fu=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=2e-5)

train_num=len(x_train)//batch_size

for epoch in range(20):
    train_loss=0
    for i in range(train_num):
        start=i*batch_size#当前批次的起始索引
        end=(i+1)*batch_size#当前批次的结束索引
        train_batch=torch.tensor(x_train[start:end]).to(device)
        label_batch=torch.tensor(y_train_label[start:end]).to(device)

        #前向传播
        pred=model(train_batch)
        loss=loss_fu(pred,label_batch)

        #反向传播
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss+=loss.item()#累加当前批次的损失

    #计算并打印损失
    train_loss/=train_num
    accuracy=(pred.argmax(1) == label_batch).type(torch.float32).sum().item()/batch_size#计算准确率
    print("epoch:",epoch+1,"train_loss:",round(train_loss,2),"accuracy:",round(accuracy,2))
```
